```python
import folium
from folium import plugins

# 시군구별 집계
district_counts = df_address['시군구명'].value_counts()

# 법정동별 집계
dong_counts = df_address['법정동명'].value_counts()

# 안산시 중심 좌표 (상록구와 단원구 중심)
ansan_center = [37.3219, 126.8309]
```

```python
import numpy as np

m = folium.Map(
    location=ansan_center,
    zoom_start=12,
    tiles='OpenStreetMap'
)

# 상록구와 단원구의 대략적인 좌표
district_coords = {
    '안산시 상록구': [37.2967, 126.8664],
    '안산시 단원구': [37.3158, 126.7932]
}

colors = ['red', 'blue']

for i, (district, count) in enumerate(district_counts.items()):
    if district in district_coords:
        folium.CircleMarker(
            location=district_coords[district],
            radius=15,  # 일정한 크기
            popup=f"{district}<br>데이터 수: {count:,}개",
            color=colors[i % len(colors)],
            fill=True,
            fillColor=colors[i % len(colors)],
            fillOpacity=0.6,
            weight=2
        ).add_to(m)

# 실제 환경에서는 지오코딩 API를 사용하거나 좌표 데이터가 필요
heat_data = []

# 법정동별 대략적인 좌표
dong_coords = {
    '일동': [37.2967, 126.8664],
    '사동': [37.2856, 126.8512],
    '본오동': [37.3123, 126.8445],
    '월피동': [37.2789, 126.8234],
    '성포동': [37.3234, 126.7823],
    '화정동': [37.3158, 126.7932],
    '선부동': [37.3289, 126.7712],
    '신길동': [37.3045, 126.8156],
    '목내동': [37.2934, 126.8789],
    '팔곡동': [37.2823, 126.8623]
}

# 히트맵 데이터 생성
for dong, count in dong_counts.head(10).items():
    if dong in dong_coords:
        lat, lon = dong_coords[dong]
        # 데이터 수에 비례한 강도로 여러 포인트 추가
        intensity = min(count / 100, 10)
        for _ in range(int(intensity)):
            # 약간의 랜덤 오프셋 추가
            lat_offset = lat + np.random.normal(0, 0.003)
            lon_offset = lon + np.random.normal(0, 0.003)
            heat_data.append([lat_offset, lon_offset, intensity])

if heat_data:
    plugins.HeatMap(
        heat_data,
        radius=15,
        blur=10,
        max_zoom=1,
        name="법정동별 밀도"
    ).add_to(m)

for dong, count in dong_counts.head(10).items():
    if dong in dong_coords:
        folium.Marker(
            location=dong_coords[dong],
            popup=f"{dong}<br>데이터 수: {count:,}개",
            icon=folium.Icon(color='green', icon='info-sign'),
            tooltip=f"{dong} ({count:,}개)"
        ).add_to(m)

# 범례 추가
legend_html = '''
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: 200px; height: 120px; 
            background-color: white; border:2px solid grey; z-index:9999; 
            font-size:14px; padding: 10px">
<h4>안산시 주소 데이터</h4>
<p><i class="fa fa-circle" style="color:red"></i> 상록구</p>
<p><i class="fa fa-circle" style="color:blue"></i> 단원구</p>
<p><i class="fa fa-map-marker" style="color:green"></i> 법정동</p>
<p>원 크기: 데이터 수 비례</p>
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# 레이어 컨트롤 추가
folium.LayerControl().add_to(m)

# 지도 저장
m.save('ansan_address_map.html')
print("\n지도가 'ansan_address_map.html' 파일로 저장되었습니다.")
```

### df_food 동 병합을 위한 전처리

```python
# 상세지역에서 '동' 추출 및 전처리
df_food['동'] = df_food['상세지역'].str.extract(r'(\w+\d*동)')

# 숫자가 포함된 동을 기본 동명으로 통일 (예: 고잔1동 → 고잔동)
df_food['동'] = df_food['동'].str.replace(r'(\w+)\d+동', r'\1동', regex=True)
```

```python
# 동 컬럼의 NaN 값을 '기타'로 대체
df_food['동'] = df_food['동'].fillna('기타')
```

```python
df_food['동'].unique()
```

```python
# 동별 수거 횟수 및 총 무게 계산
dong_weight_stats = df_food[df_food['동'] != '기타'].groupby('동').agg({
'무게': ['sum'] # sum: 총무게
}).round(2)

# 컬럼명 변경
dong_weight_stats.columns = ['음식물 폐기물 총무게']

# 총무게를 톤 단위로 변환
dong_weight_stats['총무게_톤'] = (dong_weight_stats['음식물 폐기물 총무게'] / 1000).round(2)

# 수거횟수 기준으로 내림차순 정렬
dong_weight_stats.reset_index()
```

### df_address 동 병합을 위한 전처리

```python
# 행정동명에서 '동' 추출 및 전처리
df_address['동'] = df_address['행정동명'].str.extract(r'(\w+\d*동)')

# 숫자가 포함된 동을 기본 동명으로 통일 (예: 선부3동 → 선부동)
df_address['동'] = df_address['동'].str.replace(r'(\w+)\d+동', r'\1동', regex=True)

# 동 컬럼의 NaN 값을 '기타'로 대체
df_address['동'] = df_address['동'].fillna('기타')
```

```python
# df_address의 동별로 그룹화하여 개수 계산
address_dong_counts = df_address.groupby('동').size().reset_index(name='대형폐기물 관리주소 개수')

address_dong_counts
```

### dong_weight_stats 과 address_dong_counts 병합

```python
# dong_weight_stats와 address_dong_counts 병합
merged_dong_data = pd.merge(
    dong_weight_stats[['음식물 폐기물 총무게']].reset_index(), 
    address_dong_counts, 
    on='동', 
    how='outer'
).fillna(0)

# 필요한 동 통합
merged_dong_data['동'] = merged_dong_data['동'].replace({
    '신길동': '원곡동',
    '원곡본동': '원곡동'
})

merged_dong_data['동'] = merged_dong_data['동'].replace({
    '해양동': '사동',
    '사이동': '사동',
    '중앙동': '고잔동'
})

# 통합된 동별로 데이터 재집계
merged_dong_data = merged_dong_data.groupby('동').agg({
    '음식물 폐기물 총무게': 'sum',
    '대형폐기물 관리주소 개수': 'sum'
}).reset_index()

# 대부동의 음식물 폐기물 총무게 설정
merged_dong_data.loc[merged_dong_data['동'] == '대부동', '음식물 폐기물 총무게'] = 5703940 # 자료조사

# 부곡동과 일동의 음식물 폐기물 총무게 설정
merged_dong_data.loc[merged_dong_data['동'] == '부곡동', '음식물 폐기물 총무게'] = 1739120.8 # 평균값
merged_dong_data.loc[merged_dong_data['동'] == '일동', '음식물 폐기물 총무게'] = 1739120.8 # 평균값

# 백운동, 사이동, 중앙동, 해양동의 대형폐기물 관리주소 개수 설정
for dong in ['백운동', '사이동', '중앙동', '해양동']:
    merged_dong_data.loc[merged_dong_data['동'] == dong, '대형폐기물 관리주소 개수'] = 106.7 # 평균갑


merged_dong_data
```
