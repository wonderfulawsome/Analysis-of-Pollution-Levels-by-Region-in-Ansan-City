```python
# 업종명별 개수 집계
industry_counts = df_self_check['업종명'].value_counts().head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
bars = plt.bar(industry_counts.index, industry_counts.values, color=plt.cm.Pastel2(range(len(industry_counts))))

# 막대 위에 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5, f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.title('업종명별 배출자율점검업소 지정 현황', fontsize=15, fontweight='bold', pad=15)
plt.xlabel('업종명', fontsize=12, fontweight='bold')
plt.ylabel('업소 수', fontsize=12, fontweight='bold')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_self_check['동'] = df_self_check['사업장 지번주소'].str.extract(r'(\S+동)')
df_self_check['동'] = df_self_check['동'].str.replace(r'(\w+)\d+동', r'\1동', regex=True)

# 대기종별구분명과 폐수종별구분명 별 사업장 지번주소 집계
air_group = df_self_check.groupby('대기종별구분명')['동'].unique()
wastewater_group = df_self_check.groupby('폐수종별구분명')['동'].unique()

print(air_group.head())
print(wastewater_group.head())
```

```python
# 집계년도별 배출자율점검업소 현황 시계열 시각화
year_counts = df_self_check['집계년도'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(year_counts.index.values, year_counts.values, marker='o', linewidth=2, markersize=8, color='steelblue')

# 데이터 포인트에 값 표시
for x, y in zip(year_counts.index.values, year_counts.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontweight='bold')

plt.title('연도별 배출자율점검업소 지정 현황', fontsize=15, fontweight='bold', pad=15)
plt.xlabel('집계년도', fontsize=12, fontweight='bold')
plt.ylabel('업소 수', fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 동명 통합 규칙 적용
df_self_check['동'] = df_self_check['동'].replace({
    '목내동': '초지동',
    '성곡동': '초지동',
    '원시동': '초지동',
    '사사동': '반월동',
    '신길동': '원곡동'
})
```

```python
# 가중 평균 점수 - 등급을 역순으로 변환 
air_dong_score = df_self_check.groupby('동').agg({
    '대기종별구분명': ['mean', 'count']
}).round(2)
air_dong_score.columns = ['대기_평균등급', '대기_업소수']

# 등급을 역순으로 변환 (1등급=높은 오염도 → 5점, 5등급=낮은 오염도 → 1점)
air_dong_score['대기_위험도'] = (6 - air_dong_score['대기_평균등급'])
air_dong_score['대기오염 종합점수'] = air_dong_score['대기_위험도'] * air_dong_score['대기_업소수']

wastewater_dong_score = df_self_check.groupby('동').agg({
    '폐수종별구분명': ['mean', 'count']
}).round(2)
wastewater_dong_score.columns = ['폐수_평균등급', '폐수_업소수']

# 등급을 역순으로 변환
wastewater_dong_score['폐수_위험도'] = (6 - wastewater_dong_score['폐수_평균등급'])
wastewater_dong_score['폐수배출 종합점수'] = wastewater_dong_score['폐수_위험도'] * wastewater_dong_score['폐수_업소수']

wastewater_dong_score=wastewater_dong_score.fillna(0)
air_dong_score=air_dong_score.fillna(0)
```

### 데이터 병합

```python
merged_data = pd.read_csv('merged_data.csv', encoding='utf-8-sig')
```

```python
# wastewater_dong_score에서 필요한 컬럼만 선택
wastewater_merge = wastewater_dong_score[['폐수배출 종합점수']].reset_index()

# air_dong_score에서 필요한 컬럼만 선택  
air_merge = air_dong_score[['대기오염 종합점수']].reset_index()
```

```python
# merged_data와 폐수배출 종합점수 병합
merged_data = pd.merge(
    merged_data, 
    wastewater_merge, 
    on='동', 
    how='left'
).fillna(0)

# merged_data와 대기오염 종합점수 병합
merged_data = pd.merge(
    merged_data, 
    air_merge, 
    on='동', 
    how='left'
).fillna(0)

merged_data
```

```python
# final_merged_data를 CSV 파일로 저장
merged_data.to_csv('merged_data.csv', encoding='utf-8-sig', index=False)
```
